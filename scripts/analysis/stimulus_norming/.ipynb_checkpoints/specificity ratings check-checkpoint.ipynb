```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
material = pd.read_excel('U:\ownCloud\projects\FakeDiamond\stimuli\specificity\\archive\\stimuli_specificity_complete.xlsx')
material[['specificity','word1','word2']]
material_mid = material[material.specificity == 'mid'].word1 + ' ' + material[material.specificity == 'mid'].word2
material_mid = material_mid.tolist()
material_low = material[material.specificity == 'low'].word2.tolist()
material_high = material[material.specificity == 'high'].word2.tolist()
```

```python
data = pd.read_csv('U:\ownCloud\projects\FakeDiamond\scripts\\analysis\stimulus_norming\data\data_exp_135241-v8_task-f6xa.csv')
data = data.rename(columns={'Participant External Session ID':'ptp','Display':'display','Spreadsheet: option1':'option1','Spreadsheet: option2':'option2','Spreadsheet: correct_answer':'goal','Correct':'hit','Reaction Time':'RT','Response Type':'response_type'})
data = data[['ptp','display','option1','option2','goal','hit','RT','response_type']]
data = data[(data.display == 'trial_experimental') & (data.response_type == 'response')]
data = data.replace({'ptp':dict(zip(data.ptp.unique(), range(1,len(data.ptp.unique())+1)))})

condition = []
rows_to_delete = []
for i, row in data.iterrows():
    if (row.option1 in material_low) & (row.option2 in material_mid):
        condition.append('low-mid')
    elif (row.option1 in material_mid) & (row.option2 in material_high):
        condition.append('mid-high')
    elif (row.option1 in material_low) & (row.option2 in material_high):
        condition.append('low-high')
    elif (row.option1 in material_mid) & (row.option2 in material_low):
        condition.append('low-mid')
    elif (row.option1 in material_high) & (row.option2 in material_mid):
        condition.append('mid-high')
    elif (row.option1 in material_high) & (row.option2 in material_low):
        condition.append('low-high')
    else:
        rows_to_delete.append(i)

data = data.drop(rows_to_delete)
data['condition'] = condition
data
```

First remove outliers

```python
data = data[np.abs(stats.zscore(data.RT)) < 2.5]
```

Plot histograms of different conditions

```python
sns.histplot(data=data, x='RT', hue='condition', element='step')
```

```python
columns = ['subject','condition','hit_rate','fa_rate','dprime','drop']
data_organised = pd.DataFrame(columns=columns)
dicts = []
for subject in data.ptp.unique():
    data_subject = data[data.ptp == subject]
    for condition in data_subject.condition.unique():
        # calculate hit rate and false alarm rate
        data_subject_condition = data_subject[data_subject.condition == condition]
        trial_count_subject_condition = len(data_subject_condition)
        trial_correct_subject_condition = data_subject_condition.hit.sum()
        hit_rate = trial_correct_subject_condition / trial_count_subject_condition
        fa_rate = (trial_count_subject_condition - trial_correct_subject_condition) / trial_count_subject_condition

        # calculate dprime
        dprime = stats.norm.ppf(hit_rate) - stats.norm.ppf(fa_rate)
        
        if hit_rate < 0.85:
            drop = True
        else: 
            drop = False
            
        row_values = [subject, condition, hit_rate, fa_rate, dprime, drop]
        dicts.append(row_values)

data_organised = pd.DataFrame.from_dict(dicts)
data_organised.columns = columns
data_organised = data_organised[~data_organised.isin([np.nan, np.inf, -np.inf]).any(1)]
data_organised = data_organised[~data_organised['drop']]
data_organised
```

```python
ax = sns.barplot(data=data_organised, x='condition', y='dprime')
sns.stripplot(data=data_organised, x='condition', y='dprime', dodge=True, alpha=0.75, edgecolor='gray', linewidth=1, jitter=True)
```
