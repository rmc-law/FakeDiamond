Find phrases whose mean is greater than 4 (concrete) SD is less than 1 (high agreement across raters). 

```python
ratings_concrete = ratings_all.query('Bigram == 1 & Conc_SD < 1 & Conc_M >= 4')
# ratings_concrete.reset_index(inplace=True)
ratings_concrete
```

```python
# ratings_concrete['word1'] = ''
ratings_concrete[['word1', 'word2']] = ratings_concrete['Word'].str.split(' ', 1, expand=True)
```

Implement Excel's VLOOKUP function to find constituent words' concreteness

```python
ratings_concrete['word1.Conc_M'] = ''
ratings_concrete.set_index('word1', inplace=True)
ratings_concrete['word1.Conc_M'] = ratings_concrete.index.map(ratings_all['Conc_M'])
# ratings_concrete.rename(columns={'CNC.Brysbaert': 'CNC.Brysbaert.M'}, inplace=True)
ratings_concrete
```
