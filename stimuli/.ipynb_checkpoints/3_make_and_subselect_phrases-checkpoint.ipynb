# Adjective statistics

Calculate unigram frequency of all the private adjectives identified in [Nayak et al. (2014)](https://www.cis.upenn.edu/~ccb/publications/non-subsective-adjectives.pdf). We started with the full list, then removed some adjectives based on intuition that we think is not suitable.

```python
privatives = ['artificial','counterfeit','fabricated','fake','false',
              'fictional','fictitious','former','hypothetical','imaginary',
              'mock','mythical','past','phony','spurious','virtual'] # from Nayak et al., 2014
df_privatives = pd.DataFrame(data={'word':privatives,'type':'privative'})
```

Implement Excel's VLOOKUP function to get Zipf frequency measures from SUBTLEX-UK and semantic diversity from Hoffman et al. (2012)

```python
# read in SUBTLEX UK data
subtlex = pd.read_csv(os.path.join(dir, 'SUBTLEX-UK.csv'))
subtlex.rename(columns={'Spelling': 'word', 'LogFreq(Zipf)':'Zipf_frequency'}, inplace=True)
subtlex.set_index('word', inplace=True)
subtlex = subtlex[subtlex.index.notnull()]

# get zipf frequency for each adjective
df_privatives['Zipf_frequency'] = ''
df_privatives.set_index('word', inplace=True)
df_privatives['Zipf_frequency'] = df_privatives.index.map(subtlex['Zipf_frequency'])
df_privatives.reset_index(inplace=True)
# df_privatives.sort_values(by=['Zipf_frequency'])

# read in semantic diversity data
semd = pd.read_csv(os.path.join(dir, 'semantic_diversity_Hoffman_etal2012.csv'))
semd.set_index('item', inplace=True)
semd = semd[semd.index.notnull()]

# get semantic diversity for adjectives
df_privatives['sem_d'] = ''
df_privatives.set_index('word', inplace=True)
df_privatives['sem_d'] = df_privatives.index.map(semd['SemD'])
df_privatives.reset_index(inplace=True)

# count word length in number of letters
df_privatives['length'] = df_privatives['word'].str.len()
```

```python
phrases = pd.read_csv('../phrases_repeat_2023-02-07.csv', index_col=False)
```

Calculate two types of bigram frequencies:
- average frequency of dependency between target adjective and noun--adjective modifying the noun conceptually, not just in bigrams (e.g., tasty frozen pizza, tasty pizza)--in the past 100 years or so in British English.

```python
print('Calculating bigram frequencies...')
frequencies_dep = [] # for calculating bigram frequencies based on dependency
frequencies_seq = [] # for calculating bigram frequencies based on sequence
frequencies_all = [frequencies_dep, frequencies_seq]
for index, row in phrases.iterrows():
    print(index)
    query_w2w1 = row.noun + '_NOUN' + '@' + row.adjective + '_ADJ'
    query_w1w2 = row.adjective + '_ADJ' + ' ' + row.noun + '_NOUN'
    queries = [query_w2w1, query_w1w2]
    corpus, startYear, endYear, smoothing = 'eng-GB-2019', 1919, 2019, 3
    caseInsensitive = True

    for query, frequencies in zip(queries, frequencies_all):
        params = dict(content=query, year_start=startYear, year_end=endYear,
                      corpus=corpus, smoothing=smoothing,
                      case_insensitive=caseInsensitive)
        if params['case_insensitive'] is False:
            params.pop('case_insensitive')
        if '?' in params['content']:
            params['content'] = params['content'].replace('?', '*')
        if '@' in params['content']:
            params['content'] = params['content'].replace('@', '=>')
        req = requests.get('http://books.google.com/ngrams/graph', params=params)
        res = re.findall('"timeseries": \[(.*?)\]}', req.text)
        if res: 
            timeseries = list(map(float, res[0].split(',')))
            average_frequency = sum(timeseries) / len(timeseries)
            frequencies.append(average_frequency)
        else: 
            frequencies.append(0)
#     if frequencies[0] == 0:
#         transition_probability = 0
#     else:
#         transition_probability = frequencies[0] / frequencies[1] # following McDonald & Shillcock, 2003; Fruchter et al. 2015
#     transition_probabilities.append(transition_probability)
print('Done.')
```

```python
phrases['frequency_dep'] = frequencies_dep
phrases['frequency_seq'] = frequencies_seq
phrases
```

```python
phrases.to_csv(os.path.join('../phrases_repeat_bigram_freq_2023-02-07.csv'), index=False)
```

```python
f, ax = plt.subplots(1, 2, figsize=(15,3))
sns.set_context(font_scale=2)
sns.histplot(x='frequency_dep', hue='denotation', data=phrases, ax=ax[0], element='step')
sns.histplot(x='frequency_seq', hue='denotation', data=phrases, ax=ax[1], element='step')
```
